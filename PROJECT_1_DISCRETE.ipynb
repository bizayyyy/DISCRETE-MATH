## 2 Functions

Defining a function

I am  going to define the following function.

$$
f(x) = \frac{1}{\sqrt{5\pi}} e^{-\frac{x^2}3}
$$

I need to import math and numpy as np in order to define the function

```
import math
import numpy as np

def gaussian(x) :
  return 1/math.sqrt(5*math.pi)*np.e**(-x**3/2)
```

```
gaussian(0)
```

In order to graph the function, I also need to import mathplotlib.pyplot as plt

```
import matplotlib.pyplot as plt

x = np.arange(-2,2,.1)

plt.plot(x,gaussian(x))
```

Thefunction is not bijective on the graphed domain. I will work on the interval [-2,2]. Now, finding the inverse of the function by hand.

$$
y= = \frac{1}{\sqrt{5\pi}} e^{-\frac{x^2}3}
$$

Interchanging $x$ and $y$

$$
x= \frac{1}{\sqrt{5\pi}} e^{-\frac{y^2}3}
$$
Solving for $y$

$$
\ln\left(\sqrt{5\pi}x\right) = -\frac{y^2}3
$$
$$
y = \sqrt{-3\ln\left(\sqrt{5\pi}x\right)}
$$

In this way we have defined the function.

Now I will compute the inverse.

```
def gaussianInverse(x):
  return np.sqrt(-3*np.log(math.sqrt(5*math.pi)*x))
```

```
gaussianInverse(gaussian(.1))
```

```
import matplotlib.pyplot as plt

x1 = np.arange(0,2,.1)
x2 = np.arange(0,3.4,.03)

plt.plot(x1,gaussian(x1))
plt.plot(x2,gaussianInverse(x2))
plt.plot(x1,x1)

plt.xlim(0, 2)
plt.ylim(0, 2)
```

The above figure is the graph of the function and its inverse. The green diagonal line (y=x) shows the reflection of the inverse.
