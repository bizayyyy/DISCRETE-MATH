## PART 1: Probability

I am going to play with deck of cards. The probability of getting a first card Ace is 1/13. The probability of getting a Queen is 1/13. The probability of neither is 11/13. 

```
p= [1/13,1/13,11/13]
sum(p)
```

Now I am going to create a compound event. In this case, compound event is getting either Ace or a Queen. I am going to call this draw ***R***

```
ProbabilityR = 0

for i in [0,1]:
  ProbabilityR = ProbabilityR + p[i]

ProbabilityR
```

Now I am going to calculate the probability of union and intersection of these events.

The probability of getting Ace and either Ace or Queen in two draws:
$$P(A\cap R)$$

```
intersection = p[0]*ProbabilityR
intersection
```

The probability of getting ace in second draw given that you got either Ace or Queen in first draw:
$$
P(A|R)=\frac{P(A\cap R)}{P(R)}
$$

```
RGA = intersection/ProbabilityR
RGA
```

Now, the probability of getting Ace or Queen in second draw given that you got Ace on first draw:
$$
P(R|A)=\frac{P(R\cap A)}{P(A)}
$$

```
AGR = intersection/p[0]
AGR
```

Apple Baye's Theorem

According to this theorem, 

$$
P(A|R) = \frac{P(A)P(R|A)}{P(R)}
$$


```
RGA ==(p[0]* AGR)/ProbabilityR
```

For the simulation part, I will be computing the frequency counts. So, I will multiply the probability by 13.

```
p13 = [13*i for i in p]
p13
```

Now I am going to simulate 10 draws and see if the probability matches with the random draws. For this I am going to use this specific library.

```
from random import choices


draw = choices(['Ace','Queen','Other'],p13,k=10)
print(draw)

countAce = 0;
countQueen = 0;
countOthers = 0;

for i in draw:
  if i == 'Ace':
    countAce += 1
  if i == 'Queen':
    countQueen += 1
  if i == 'Other':
    countOthers += 1

estimateProb = [countAce/len(draw), countQueen/len(draw), countOthers/len(draw)]
estimateProb
```
